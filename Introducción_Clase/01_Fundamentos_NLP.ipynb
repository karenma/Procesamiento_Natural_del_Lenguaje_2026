## 1. ¿Qué es el Procesamiento de Lenguaje Natural?

El **Procesamiento de Lenguaje Natural** (NLP, por sus siglas en inglés: *Natural Language Processing*) es una rama de la inteligencia artificial que se ocupa de la interacción entre computadoras y el lenguaje humano.

Su objetivo es permitir que las máquinas **lean, entiendan, interpreten y generen** lenguaje humano de una manera que sea útil.

>  **En términos simples:** NLP es la tecnología detrás de que tu asistente de voz te entienda, de que Google traduzca texto, o de que un chatbot responda tus preguntas.

### Historia breve del NLP

| Época | Hito |
|---|---|
| **1950s** | Test de Turing — Alan Turing propone si las máquinas pueden pensar |
| **1960s** | ELIZA — primer chatbot basado en reglas (MIT) |
| **1980s** | Sistemas basados en reglas gramaticales formales |
| **1990s** | Machine Learning estadístico: HMM, Naive Bayes |
| **2000s** | SVM, CRF para etiquetado de secuencias |
| **2013** | Word2Vec — representaciones vectoriales densas |
| **2017** | Transformer — *Attention Is All You Need* (Vaswani et al.) |
| **2018** | BERT — Google revoluciona el NLP con modelos pre-entrenados |
| **2020+** | GPT-3, ChatGPT, LLMs — lenguaje generativo a gran escala |



### Ramas del NLP

El NLP se conecta con múltiples disciplinas:

```
         Lingüística
              │
   Psicología─┼─Ciencias Cognitivas
              │
         NLP  ◄──── Inteligencia Artificial
              │
    Estadística┼─Machine Learning
              │
        Ingeniería de Software
```

## 3. Conceptos Clave 

### 📦 Token

Un **token** es la unidad mínima de análisis en NLP. Puede ser:
- Una palabra: `"hola"`, `"mundo"`
- Un subword: `"##ción"` (BERT)
- Un carácter: `'h'`, `'o'`, `'l'`, `'a'`
- Un símbolo de puntuación: `','`, `'.'`

**Ejemplo de tokenización:**
```
Oración: "El gato come pescado."
Tokens:  ["El", "gato", "come", "pescado", "."]
```

La tokenización es el **primer paso** en cualquier pipeline NLP.


### 📚 Corpus

Un **corpus** (plural: *corpora*) es una colección estructurada de textos utilizada para entrenar, evaluar o estudiar modelos NLP.

#### Tipos de corpus:

| Tipo | Descripción | Ejemplo |
|---|---|---|
| **Monolingüe** | En un solo idioma | Wikipedia en español |
| **Bilingüe** | Dos idiomas alineados | Europarl (EN-ES) |
| **Anotado** | Con etiquetas POS, NER, etc. | Penn Treebank |
| **De dominio** | Especializado en un tema | PubMed (biomédico) |
| **De sentimientos** | Con polaridad etiquetada | IMDB Reviews |

#### Corpora famosos:
- **Brown Corpus** (1961) — primer corpus computacional del inglés
- **Penn Treebank** — análisis sintáctico
- **Wikipedia Dumps** — corpus masivo en 300+ idiomas
- **Common Crawl** — rastreo web, usado para entrenar GPT
- **IMDB** — 50,000 reseñas de cine etiquetadas
- **SQUAD** — preguntas y respuestas sobre Wikipedia



### 📖 Vocabulario

El **vocabulario** (o *lexicón*) es el conjunto de todos los tokens únicos presentes en un corpus.

```
Corpus: ["el perro corre", "el gato duerme", "el perro duerme"]
Vocabulario: {"el", "perro", "corre", "gato", "duerme"}
Tamaño: |V| = 5
```

#### Conceptos relacionados:

| Concepto | Definición |
|---|---|
| **Type** | Token único en el vocabulario |
| **Token** | Instancia de un type en el corpus |
| **OOV** | Out-of-Vocabulary — palabras no vistas durante entrenamiento |
| **Ley de Zipf** | La frecuencia de una palabra es inversamente proporcional a su rango |

#### La Ley de Zipf

En cualquier idioma natural, pocas palabras son muy frecuentes y la mayoría son raras. Esto es la **Ley de Zipf**:

$$f(r) \propto \frac{1}{r}$$

donde $f$ es la frecuencia y $r$ es el rango (posición en el ranking).

> Las 100 palabras más frecuentes del inglés cubren ~50% de cualquier texto. Las stopwords son un ejemplo de esto.


## 4. Embeddings: Intuición y Uso 
### El Problema de Representar Palabras

Las computadoras no entienden palabras directamente — necesitamos convertirlas en números.

#### Representación One-Hot (ingenua)

```
Vocabulario: ["rey", "reina", "hombre", "mujer"]

rey   → [1, 0, 0, 0]
reina → [0, 1, 0, 0]
hombre→ [0, 0, 1, 0]
mujer → [0, 0, 0, 1]
```

**Problemas:** vectores dispersos, no captura similaridad semántica.


### Word Embeddings: La Solución

Un **embedding** es una representación densa de baja dimensión que **captura significado semántico**.

```
rey   → [0.91, 0.32, -0.45, ..., 0.12]  (300 dimensiones)
reina → [0.88, 0.29, -0.41, ..., 0.09]
hombre→ [0.45, 0.68, -0.21, ..., 0.33]
mujer → [0.42, 0.71, -0.18, ..., 0.30]
```

#### La famosa analogía de Word2Vec:

$$\vec{rey} - \vec{hombre} + \vec{mujer} \approx \vec{reina}$$

Esto significa que los embeddings **aprenden relaciones semánticas y sintácticas**.



### Tipos de Embeddings

| Modelo | Año | Característica |
|---|---|---|
| **Word2Vec** | 2013 | CBOW y Skip-gram, Google |
| **GloVe** | 2014 | Factorización de co-ocurrencias, Stanford |
| **FastText** | 2016 | Sub-word tokens, maneja OOV, Facebook |
| **ELMo** | 2018 | Embeddings contextuales (LSTM) |
| **BERT** | 2018 | Bidireccional, contexto completo |
| **Sentence-BERT** | 2019 | Embeddings a nivel oración |



### Visualización de Embeddings

Usando técnicas de reducción de dimensionalidad como **t-SNE** o **PCA**, podemos visualizar la geometría semántica:

```
        reina ●                    ● princesa
              
    rey ●                    ● príncipe
    
    mujer ●                    ● chica
    
    hombre ●                   ● chico
```

Las palabras con significados similares aparecen **agrupadas** en el espacio vectorial.

### ¿Para qué sirven los Embeddings?

- **Búsqueda semántica**: encontrar documentos relevantes aunque no compartan palabras exactas
- **Clasificación de texto**:features ricos para modelos de ML
- **Traducción automática**: alinear espacios semánticos entre idiomas
- **Clustering**: agrupar documentos por tema
- **RAG**: Retrieval Augmented Generation en LLMs

## 5. Pipeline Típico de un Proyecto NLP 

Un proyecto NLP sigue una secuencia de etapas bien definidas. Este es el **pipeline estándar**:

```
┌─────────────────────────────────────────────────────────────┐
│                   PIPELINE NLP COMPLETO                     │
├─────────────────────────────────────────────────────────────┤
│                                                             │
│  1. DATOS CRUDOS                                            │
│     └─ Texto sin procesar (tweets, PDFs, HTML, etc.)        │
│                    ↓                                        │
│  2. ADQUISICIÓN Y LIMPIEZA                                  │
│     └─ Scraping, OCR, parsing, eliminación de ruido         │
│                    ↓                                        │
│  3. PREPROCESAMIENTO                                        │
│     ├─ Tokenización                                         │
│     ├─ Normalización (minúsculas, acentos)                  │
│     ├─ Eliminación de stopwords                             │
│     ├─ Stemming / Lematización                              │
│     └─ Manejo de OOV                                        │
│                    ↓                                        │
│  4. REPRESENTACIÓN / FEATURES                               │
│     ├─ Bag of Words / TF-IDF                                │
│     ├─ Word Embeddings (Word2Vec, GloVe)                    │
│     └─ Contextual Embeddings (BERT, RoBERTa)                │
│                    ↓                                        │
│  5. MODELADO                                                │
│     ├─ Modelos clásicos (Naive Bayes, SVM, LR)              │
│     ├─ Deep Learning (LSTM, CNN)                            │
│     └─ Transformers (BERT, GPT)                             │
│                    ↓                                        │
│  6. EVALUACIÓN                                              │
│     ├─ Accuracy, F1, Precision, Recall                      │
│     ├─ BLEU, ROUGE (generación de texto)                    │
│     └─ Análisis de errores                                  │
│                    ↓                                        │
│  7. DESPLIEGUE                                              │
│     ├─ API REST (FastAPI, Flask)                            │
│     ├─ Interfaz web (Gradio, Streamlit)                     │
│     └─ Monitoreo en producción                              │
└─────────────────────────────────────────────────────────────┘
```

### Ejemplo: Pipeline de Análisis de Sentimientos

| Etapa | Input | Output |
|---|---|---|
| Datos crudos | `"¡Esta película fue INCREÍBLE!! 😍"` | texto raw |
| Limpieza | Eliminar emojis, normalizar | `"Esta película fue increíble"` |
| Tokenización | Separar tokens | `["esta", "película", "fue", "increíble"]` |
| Stop words | Eliminar palabras vacías | `["película", "increíble"]` |
| Vectorización | TF-IDF / Embedding | `[0.0, 0.71, 0.0, ..., 0.89]` |
| Modelo | Clasificador entrenado | `POSITIVO (0.94)` |


### Consideraciones importantes

1. **No todos los pasos son necesarios siempre**:  depende de la tarea y los datos
2. **Los modelos modernos (BERT, GPT) reducen el preprocesamiento**:  aprenden directamente del texto
3. **La calidad de los datos es más importante que el modelo**:  "Garbage in, garbage out"
4. **El pipeline debe ser reproducible**:  usar pipelines de scikit-learn o spaCy
