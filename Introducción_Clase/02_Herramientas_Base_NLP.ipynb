## 3. Módulo  Expresiones Regulares (`re`)

El módulo `re` de Python permite trabajar con **expresiones regulares**: patrones para buscar, extraer y transformar texto.

[Documentación](https://docs.python.org/3/library/re.html  )

[Herramienta de prueba](https://regex101.com/)

### Metacaracteres principales

| Patrón | Descripción | Ejemplo |
|---|---|---|
| `.` | Cualquier carácter (excepto newline) | `a.c` → `abc`, `a1c` |
| `^` | Inicio de cadena | `^Hola` → empieza con "Hola" |
| `$` | Fin de cadena | `mundo$` → termina con "mundo" |
| `*` | 0 o más repeticiones | `ab*c` → `ac`, `abc`, `abbc` |
| `+` | 1 o más repeticiones | `ab+c` → `abc`, `abbc` |
| `?` | 0 o 1 repetición | `colou?r` → `color`, `colour` |
| `{n}` | Exactamente n repeticiones | `\d{4}` → 4 dígitos |
| `{n,m}` | Entre n y m repeticiones | `\d{2,4}` → 2 a 4 dígitos |
| `[]` | Clase de caracteres | `[aeiou]` → vocales |
| `[^]` | Negación de clase | `[^0-9]` → no dígitos |
| `\|` | Alternativa (OR) | `gato\|perro` → gato o perro |
| `()` | Grupo de captura | `(\w+)@` |

### Secuencias especiales

| Secuencia | Descripción |
|---|---|
| `\d` | Dígito `[0-9]` |
| `\D` | No dígito `[^0-9]` |
| `\w` | Carácter alfanumérico `[a-zA-Z0-9_]` |
| `\W` | No alfanumérico |
| `\s` | Espacio en blanco (espacio, tab, newline) |
| `\S` | No espacio en blanco |
| `\b` | Límite de palabra |
| `\B` | No límite de palabra |

### Funciones principales de `re`

| Función | Descripción |
|---|---|
| `re.match(pattern, string)` | Busca al inicio de la cadena |
| `re.search(pattern, string)` | Busca en cualquier posición |
| `re.findall(pattern, string)` | Retorna lista de todas las coincidencias |
| `re.finditer(pattern, string)` | Retorna iterador de Match objects |
| `re.sub(pattern, repl, string)` | Reemplaza coincidencias |
| `re.split(pattern, string)` | Divide la cadena por el patrón |
| `re.compile(pattern)` | Compila patrón para reutilización |
| `re.fullmatch(pattern, string)` | La cadena completa debe coincidir |

##### Limpieza de Texto

````python
## re.sub: Limpieza de texto 
texto_sucio = """¡¡OFERTA!!! Compra AHORA 😍🔥 el MEJOR producto.
Visita: https://www.tienda.com/oferta?id=123&ref=promo
Llama al: +52 (55) 1234-5678 o escríbenos a ventas@tienda.com
   Mucho   espacio   en   blanco   aquí   """
texto_sucio
````

````python
# Paso 1: Eliminar URLs
texto_clean = re.sub(r'https?://\S+', '[URL]', texto_sucio)
````

````python
texto_clean
````

````python
# Paso 2: Eliminar emojis (caracteres no ASCII)
texto_clean = re.sub(r'[^\x00-\x7Fáéíóúüñ¡¿ÁÉÍÓÚ \n.,!?@()+-]', '', texto_clean)
texto_clean
````

````python
# Paso 3: Eliminar números de teléfono
texto_clean = re.sub(r'[+]?[\d\s().,-]{10,}', '[TELÉFONO]', texto_clean)
texto_clean
````

````python
# Paso 4: Eliminar correos
texto_clean = re.sub(r'[\w.]+@[\w.]+\.[a-zA-Z]{2,}', '[EMAIL]', texto_clean)
texto_clean
````

````python
# Paso 5: Normalizar espacios múltiples
texto_clean = re.sub(r'\s+', ' ', texto_clean).strip()
texto_clean
````

````python
# Paso 6: Convertir a minúsculas
texto_clean = texto_clean.lower()
````

````python
# Grupos de captura
print("📌 GRUPOS DE CAPTURA: Extracción clave:valor:")
patron = r'(\w+[\w\s]*):\s*([\w\s]+)'
for match in re.finditer(patron, texto):
    clave, valor = match.group(1), match.group(2).strip()
    print(f"  Clave: '{clave}' → Valor: '{valor}'")
````

## Regex para NLP: casos de uso comunes 

````python
## Regex para NLP: casos de uso comunes 
import re

tweets = [
    "¡Me #NLP encanta el #MachineLearning! Es el futuro 🚀 @OpenAI",
    "Revisando el paper de #BERT https://arxiv.org/abs/1810.04805 cc @GoogleAI",
    "HOY APRENDÍ A USAR spaCy 🎉🎉 #NLP #Python @huggingface",
]
tweets
````

````python
#Preprocesar el texto limpiando:
for tweet in tweets:
    hashtags = re.findall(r'#\w+', tweet)
    menciones = re.findall(r'@\w+', tweet)
    emojis = re.findall(r'[^\w\sáéíóúüñ¡¿]', tweet)

    print(f'Tweet: {tweet}')
    print(f'Hashtags: {hashtags}')
    print(f'Menciones: {menciones}')
    print(f'Emojjis: {emojis}')
# Hashtags #\w+
# Menciones @\w+
# Ligas https?://\S+
# Emojis ^\w\sáéíóúüñ¡¿
````

````python
for tweet in tweets:
    limpio = re.sub(r'#\w+|@\w+|https?://\S+','',tweet)
    limpio = re.sub(r'[^\w\sáéíóúüñ¡¿]','',limpio)
    print(f'Tweet: {tweet}')
    print(f'Tweet Limpio: {limpio}')
    print()
````

````python
# Método 1: str.maketrans (muy eficiente)
tabla = str.maketrans('', '', string.punctuation + '¡¿')
limpio = texto.translate(tabla)
print(" Limpieza de puntuación:")
print(f"  Original: '{texto}'")
print(f"  Limpio:   '{limpio}'")
````

````python
# Método 2: comprensión de listas
tokens = [c for c in texto.lower() if c not in string.punctuation + '¡¿']
texto_filtrado = ''.join(tokens)
print(f"  Comprensión: '{texto_filtrado}'")
````

````python
# Uso con tokens
palabras = texto.lower().split()
palabras_limpias = [''.join(c for c in p if c not in string.punctuation + '¡¿') 
                    for p in palabras]
palabras_limpias = [p for p in palabras_limpias if p]  # eliminar vacíos
print(f"  Palabras limpias: {palabras_limpias}")
````
