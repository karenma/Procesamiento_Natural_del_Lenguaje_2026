## 0. Por que vectorizar el texto?

Los algoritmos de Machine Learning no pueden procesar texto directamente.
Necesitamos transformar el texto en **vectores numericos**.

### El problema de representacion:
```
"El gato se sienta en la alfombra"
                |
                v  vectorizar
[0, 1, 0, 1, 0, 1, 1, 1, 0, ...]  -> Vector numerico
```


## 1. Bag of Words 

### ¿Qué es Bag of Words?

**Bag of Words (BoW)** representa cada documento como un **vector de frecuencias** de palabras, ignorando el orden y la gramática.

#### Proceso de construccion:
```
1. Obtener el Vocabulario (conjunto de palabras unicas):
   V = {alfombra, amigos, corre, duerme, el, ...}

2. Cada documento se convierte en un vector de conteos:
   Doc: 'El gato duerme'
   V:    [alfombra, amigos, corre, duerme,  el, gato, ...]
   Vec:  [    0,      0,      0,      1,    1,    1,  ...]
```

#### Ejemplo
```
Doc 1: "el gato come pescado"
Doc 2: "el perro come carne"
Doc 3: "el gato juega con el perro"

Vocabulario: [carne, come, con, el, gato, juega, perro, pescado]

Doc 1: [0, 1, 0, 1, 1, 0, 0, 1]
Doc 2: [1, 1, 0, 1, 0, 0, 1, 0]
Doc 3: [0, 0, 1, 2, 1, 1, 1, 0]
```
#### Formula:
$$BoW(d, w) = count(w, d)$$

Donde $d$ es el documento y $w$ es la palabra del vocabulario.

#### Ventajas y limitaciones

| Ventajas | Limitaciones |
|---|---|
| Simple de implementar | Ignora el orden de las palabras |
| Funciona bien para clasificación | Vectores muy dispersos (sparse) |
| Interpretable | No captura semántica |
| Rápido | Las palabras frecuentes dominan |


#### Variantes de BoW:
- **Conteo binario:** 1 si aparece, 0 si no
- **Conteo de frecuencia:** numero de veces que aparece
- **Frecuencia normalizada:** frecuencia / total de palabras del doc

## Algunas Limitaciones de BoW

El modelo Bag of Words tiene importantes limitaciones:

##### 1. Pierde el orden de las palabras:
```
"El perro muerde al hombre" = "El hombre muerde al perro"
(mismo BoW, pero significados completamente diferentes)
```

##### 2. No captura semantica:
```
"rey" y "reina" son vectores sin relacion en BoW
"auto" y "coche" son vectores sin relacion (sinonimos)
```

##### 3. Alta dimensionalidad y esparsidad:
```
Corpus con 50,000 palabras -> vector de 50,000 dimensiones
Un documento tipico usa ~500 palabras -> 99% del vector es 0
```

##### 4. Palabras frecuentes dominan:
```
"el", "la", "de" aparecen en todos los documentos
-> Son los vectores mas importantes en BoW (pero los menos informativos!)
TF-IDF resuelve este problema.
```

##### 5. No maneja sinonimos ni polisemia:
```
Sinonimos: "auto" y "coche" -> vectores distintos
Polisemia: "banco" (financiero) y "banco" (mueble) -> mismo vector
```

## 3. TF-IDF (Concepto)

### ¿Qué es TF-IDF?

**TF-IDF** (**Term Frequency – Inverse Document Frequency**) mide la **importancia de una palabra en un documento relativa al corpus**. Penaliza palabras muy comunes y premia palabras específicas. Mejora a BoW al penalizar las palabras que aparecen en demasiados documentos.

#### Term Frequency (TF) 
Que tan frecuente es la palabra en el documento

$$TF(t, d) = \frac{\text{número de veces que } t \text{ aparece en } d \text { }(count(t,d))} {\text{total de palabras en } d}$$

#### Inverse Document Frequency (IDF)

$$IDF(t) = \log \left( \frac{N}{df(t)} \right) + 1$$

donde $N$ = número de documentos y $df(t)$ = documentos que contienen $t$. El $+1$ es suavizado (smoothing) para evitar division por cero

#### TF-IDF

$$TFIDF(t, d) = TF(t, d) \times IDF(t)$$

```
Palabra muy frecuente en doc Y muy frecuente en corpus:
  'el', 'la', 'de' -> TF alto, IDF bajo -> TF-IDF bajo (no importante)

Palabra frecuente en doc PERO rara en corpus:
  'tokenizacion', 'transformer' -> TF alto, IDF alto -> TF-IDF alto (importante!)

Palabra rara en doc Y rara en corpus:
  nombres propios raros -> TF bajo, IDF alto -> TF-IDF bajo
```

#### Interpretación

| Caso | TF | IDF | TF-IDF | Significado |
|---|---|---|---|---|
| Palabra muy frecuente en todo el corpus ("el") | Alto | Bajo | Bajo | Stopword — no discrimina |
| Palabra frecuente en UN documento | Alto | Alto | **Alto** | Relevante para ese doc |
| Palabra rara en todo el corpus | Bajo | Alto | Bajo | Ruido |
| Palabra clave específica del doc | Medio | Alto | **Alto** | Muy relevante |

#### Ejemplo de TFIDF Sin librerías

````python
import math

class TFIDF:
    def __init__(self, smooth_idf=True, norm='l2'):
        self.smooth_idf = smooth_idf
        self.norm = norm
        self.vocabulario = []
        self.word2idx = {}
        self.idf_values = {}

    def _calcular_tf(self, tokens):
        n = len(tokens)
        if n == 0: return {}
        conteo = Counter(tokens)
        return {w: c/n for w, c in conteo.items()}

    def _calcular_idf(self, corpus_tokens):
        N = len(corpus_tokens)
        df = Counter()
        for tokens in corpus_tokens:
            for w in set(tokens):
                df[w] += 1
        if self.smooth_idf:
            return {w: math.log((N + 1) / (d + 1)) + 1 for w, d in df.items()}
        else:
            return {w: math.log(N / d) + 1 for w, d in df.items()}

    def fit(self, corpus):
        corpus_tokens = [preprocesar(doc, idioma = '') for doc in corpus]
        self.vocabulario = sorted(set(w for toks in corpus_tokens for w in toks))
        self.word2idx = {w: i for i, w in enumerate(self.vocabulario)}
        self.idf_values = self._calcular_idf(corpus_tokens)
        return self

    def transform(self, corpus):
        matriz = np.zeros((len(corpus), len(self.vocabulario)))
        for i, doc in enumerate(corpus):
            tokens = preprocesar(doc, idioma = '')
            tf = self._calcular_tf(tokens)
            for w, tf_val in tf.items():
                if w in self.word2idx:
                    j = self.word2idx[w]
                    matriz[i, j] = tf_val * self.idf_values.get(w, 0)
        # Normalizacion L2
        if self.norm == 'l2':
            normas = np.linalg.norm(matriz, axis=1, keepdims=True)
            normas = np.where(normas == 0, 1, normas)
            matriz = matriz / normas
        return matriz

    def fit_transform(self, corpus):
        return self.fit(corpus).transform(corpus)
````
